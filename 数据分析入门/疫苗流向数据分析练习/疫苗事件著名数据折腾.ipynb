```python
create_co_cnt = df['create_company'].value_counts()
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1)
ax.set_title('各公司产品数量', fontproperties=myfont)
ax.set_ylabel('公司', fontproperties=myfont)
ax.set_xlabel('产品数量', fontproperties=myfont)
plt.barh(range(len(create_co_cnt)), list(create_co_cnt))
plt.yticks(range(len(create_co_cnt)), list(create_co_cnt.keys()), fontproperties=myfont)
plt.show()
```
